## 12.1 Selecting Best Models using Exhaustive search

Selecting a best model by searching over a range of hyperparameters.

```python
import numpy as np
from sklearn import linear_model, datasets
from sklearn.model_selection import GridSearchCV
```

```python
# loading data

iris = datasets.load_iris()
features = iris.data
target = iris.target
```

```python
# create logistic regression

logistic = linear_model.LogisticRegression()
```

```python
# create range of candidate peanlty hyperparameter values

penalty = ['l1','l2']
```

```python
# create range of candidtae regularization hyperparameter values

C = np.logspace(0, 4, 10)
```

```python
# create dictionary hyperparameter candidates

hyperparameters = dict(C = C, penalty = penalty)
```

```python
# create grid search

gridSearch = GridSearchCV(logistic, hyperparameters, cv = 5, verbose = 0)

# verbose parameter determines amount of messages outputted during the search, 
# with 0 showing no output, and 1 to 3 outputting messages with incresed detail.
```

**C and regularization Penalty can take range of values and have to be specified prior to training.**

```python
# fit grid search
best_model = gridSearch.fit(features, target)
```

```python
# for C we define 10 possible values and l1, l2 are the values of penalty

np.logspace(0, 4, 10)
```

we train the model and evaluate using KFCV, here we have 10 values of C, 2 values of Penalty and 5 folds, they created 10 x 2 x 5 = 100 candidate models from which best model is selected. 

```python
# view best hyperparameters

print('Best Penalty:', best_model.best_estimator_.get_params()['penalty'])
print('Best C:', best_model.best_estimator_.get_params()['C'])
```

After identifying the best hyperparameters, GridSearchCV will retrain a model using best hyperparameters on the entire dataset, we can use this model to predict values like any other scikit-learn model

```python
# predict target vector

best_model.predict(features)
```

## 12.2 Selecting Best Models Using Randomized Search

Computationally cheaper method than exhaustive search

```python
# loading libraries

from scipy.stats import uniform
from sklearn import linear_model, datasets
from sklearn.model_selection import RandomizedSearchCV
```

```python
# loaing data

iris = datasets.load_iris()
features = iris.data
target = iris.target
```

```python
# creating logistic regression

logistic = linear_model.LogisticRegression()
```

```python
# creating range candidate regularization penalty hyperparameter values

penalty = ['l1','l2']
```

```python
# create distribution of candidate regularization hyperparameter values

C = uniform(loc = 0, scale = 4)
```

```python
# create hyperparameter options

hyperparameters = dict(C = C, penalty = penalty)
```

```python
# creating randomized search

randomizedSearch = RandomizedSearchCV(logistic, hyperparameters, random_state = 1, 
                                      n_iter = 100, cv = 5, verbose = 0, n_jobs = -1)
```

```python
# fit randomized search

best_model = randomizedSearch.fit(features, target)
```

```python
# here we randomly sample 10 values from a uniform distribution rangigng from 0 to 4

uniform(loc = 0, scale = 4).rvs(10)
```

```python
# view the best parameters

print('Best Penalty:', best_model.best_estimator_.get_params()['penalty'])
print('Best C:', best_model.best_estimator_.get_params()['C'])
```

```python
# predicting target vector

best_model.predict(features)
```

## 12.3 Selecting Best Models from Multiple Learning Algorithms

Selecting a best model by searching over a range of learning algorithms and their respective hyperparameters.

```python
# loading libraries

import numpy as np
from sklearn import datasets
from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import GridSearchCV
from sklearn.pipeline import Pipeline
```

```python
# set random seed

np.random.seed(0)
```

```python
# load data

iris = datasets.load_iris()
features = iris.data
target = iris.target
```

```python
# creating a pipeline

pipe = Pipeline([("classifier", RandomForestClassifier())])
```

We are defining a search space that with 2 learning algorithms:
1) LogisticRegression
2) RandomForestClassifier

defining penalty and C hyperparameters adn random forest hyperparameters

```python
# creating dictionary with candidate learning algorithms and their hyperparameters

search_space = [{"classifier": [LogisticRegression()],
                "classifier__penalty": ["l1","l2"],
                "classifier__c": np.logspace(0, 4, 10)},
               {"classifer": [RandomForestClassifier()],
               "classifier__n_estimators": [10, 100, 1000],
               "classifier__max_features": [1, 2, 3]}]
```

```python
# creating grid search

gridsearch = GridSearchCV(pipe, search_space, cv = 5, verbose = 0)
```

```python
# fit grid search

best_model = gridsearch.fit(features, target)
```

```python
# view best model

best_model.best_estimator_.get_params()["classifier"]
```

```python
# predicting target vector

best_model.predict(features)
```

## 12.4 Selecting Best Models When Preprocessing

Including a preprocessing step during model selection.

```python
# load libraries

import numpy as np
from sklearn import datasets
from sklearn.linear_model import LogisticRegression
from sklearn.model_selection import GridSearchCV
from sklearn.pipeline import Pipeline, FeatureUnion
from sklearn.decomposition import PCA
from sklearn.preprocessing import StandardScaler
```

```python
# set random seed

np.random.seed(0)
```

```python
# loading data
iris = datasets.load_iris()
features = iris.data
target = iris.target
```

FeatureUnion allows us to combine multiple preprocessing actions properly. 

Here we are combining two preprocessing steps: StandardScalar() and PCA()

```python
# creating a preprocessing object that includes Standardscaler features and PCA

preprocess = FeatureUnion([("std", StandardScaler()),("pca", PCA())])
```

```python
# creating a pipeline

pipe = Pipeline([("preprocess", preprocess),
                ("classifier", LogisticRegression())])
```

```python
# create space of candidate values

search_space = [{"preprocess__pca__n_components": [1, 2, 3],
                "classifier__penalty": ["l1","l2"],
                "classifier__C": np.logspace(0, 4, 10)}]
```

```python
# creating a grid search

clf = GridSearchCV(pipe, search_space, cv = 5, verbose = 0, n_jobs = -1)
```

```python
# fit grid search

best_model = clf.fit(features, target)
```

```python
# view best model

best_model.best_estimator_.get_params()['preprocess__pca__n_components']
```

## 12.5 Speeding up Model Selection with Parallelization

The parameter n_jobs tells number of models to train in parallel.

Using all the cores in the machine by setting n_jobs = -1.

```python
# load libraries

import numpy as np
from sklearn import linear_model, datasets
from sklearn.model_selection import GridSearchCV
```

```python
# loading data

iris = datasets.load_iris()
features = iris.data
target = iris.target
```

```python
# create logistic regression

logistic = linear_model.LogisticRegression()
```

```python
# creating range of candidate regularization penalty hyperparameter values

penalty = ["l1","l2"]
```

```python
# creating a range of candidate values for C

C = np.logspace(0, 4, 1000)
```

```python
# creating hyperparameter options

hyperparameters = dict(C = C, penalty = penalty)
```

```python
# create a grid search

gridsearch = GridSearchCV(logistic, hyperparameters, cv = 5, n_jobs = -1, verbose = 1)
```

```python
# fit grid search

best_model = gridsearch.fit(features, target)
```

```python
# creating grid search using one core

clf = GridSearchCV(logistic, hyperparameters, cv = 5, n_jobs = 1, verbose = 1)
```

```python
# fit grid search

best_model = clf.fit(features, target)
```

## 12.6 Speeding Up Model Selection Using Algorithm-Specific Methods

While using a select number of learning algorithms, we can use scikit-learn's model-specific cross-validation hyperparameter tuning, LogisticRegression is used to conduct a standard logistic regression classifier, while LogisticRegressionCV implements an efficient cross-validated logistic regression classifier:

```python
from sklearn import linear_model, datasets
```

```python
# loading dataset

iris = datasets.load_iris()
features = iris.data
target = iris.target
```

```python
# creating the cross-validate logistic regression

logit = linear_model.LogisticRegressionCV(Cs = 100)
```

```python
# train model

logit.fit(features, target)
```

The major downside of LogisticRegressionCV is that it can only search a range of values for C(not both C and penalty)

## 12.7 Evaluating Performance After Model Selection

To evaluate the performance of a model found through model selection.

```python
import numpy as np
from sklearn import linear_model, datasets
from sklearn.model_selection import GridSearchCV, cross_val_score
```

```python
# loading the dataset

iris = datasets.load_iris()
features = iris.data
target = iris.target
```

```python
# creating logistic regression

logit = linear_model.LogisticRegression()
```

```python
# creating a range of 20 candidate values for C

C = np.logspace(0, 4, 20)
```

```python
# create hyperparameter options

hyperparameters = dict(C = C)
```

```python
# create grid search
gridsearch = GridSearchCV(logit, hyperparameters, cv = 5, n_jobs = -1, verbose = 0)
```

```python
# conducting nested cross-validation and outputting the average score

cross_val_score(gridsearch, features, target).mean()
```

It is a nested cross-validation, the inner cross-validation selects the best model, while the outer cross-validation provides us with an unbaised evaluation of model's performance.

In the above scenario, GridSearchCV is our inner cross-validation which we can wrap in an outer cross-validation using cross_val_score.
