## 3.14 Grouping Rows by Time

```python
time_index = pd.date_range('06/06/2017', periods=100000, freq='30S')
# taking time index of time by a time period difference of 30 seconds 

# creating a dataframe
df = pd.DataFrame(index=time_index)

# creating a column of random variables
df['Sale Amount'] = np.random.randint(1,10,100000) 
df
```

```python
df.resample('W').sum()

# grouping the rows by weeks and finding sum of sale amount for the week
```

```python
df.resample('2W').sum()
# grouping by 2 weeks
```

```python
df.resample('M').mean()
# grouping by months and calculating means
```

```python
# resample returns the right edge of the label in the datetime column
# we can specify left edge using "label" 

df.resample('M',label='left').mean()
```

## 3.18 Concatenating DataFrames

```python
# axis=0 to concatenate along the axis

data_1 = {'id':['1','2','3'],
         'firstName':['Ben','Mick','Cal'],
         'lastName':['Stone','Stone','Stone']}
```

```python
data_1
```

```python
df_1 = pd.DataFrame(data_1, columns=['id', 'firstName', 'lastName'])
df_1
```

```python
data_2 = {'id':['4','5','6'],
         'firstName':['Ben','Ross','Emma'],
         'lastName':['Gellar','Gellar','Gellar']}
df_2 = pd.DataFrame(data_2, columns=['id', 'firstName', 'lastName'])
df_2
```

```python
pd.concat([df_1,df_2], axis = 0)
# 'axis = 0' to concatenate along rows
```

```python
pd.concat([df_1,df_2], axis = 1)
# 'axis = 1' to concatenate along columns
```

## 3.19 Merging DataFrames

merge() defaults to  Inner Join

```python
employee = {'id':['1','2','3','4'],
         'firstName':['Ben','Ross','Emma', 'Rachel'],
         'lastName':['Gellar','Gellar','Gellar','Gellar']}
emp = pd.DataFrame(employee, columns=['id', 'firstName', 'lastName'])
emp
```

```python
emp_role = {'id':['3','4','5'],
         'role':['Developer','Tester','Director']}
role = pd.DataFrame(emp_role, columns=['id', 'role'])
role
```

```python
pd.merge(emp, role, on='id')
# inner join - on
```

```python
pd.merge(emp, role, on='id', how = 'outer')
# outer join - how
```

```python
pd.merge(emp, role, on='id', how = 'left')
# left join
```

```python
pd.merge(emp, role, on='id', how = 'right')
# right join
```

```python
pd.merge(emp, role, left_on = 'id', right_on = 'id')
# outer join - how
```
