## 8.8 Isolating colors

we first have to define the colors then apply mask to the image:

```python
# loading image

image = cv2.imread("birdy.jpg")
```

```python
# converting image to HSV - hue, saturation, value

image_hsv = cv2.cvtColor(image, cv2.COLOR_BGR2HSV)
```

```python
# Defining the range of blue values in HSV

lower_blue = np.array([0, 0, 200])
upper_blue = np.array([180, 30, 255])
```

```python
# crteating mask

mask = cv2.inRange(image_hsv, lower_blue, upper_blue)
```

```python
# mask the image

img_bgr_masked = cv2.bitwise_and(image, image, mask = mask)
```

```python
# converting BGR to RGB

img_rgb = cv2.cvtColor(img_bgr_masked, cv2.COLOR_BGR2RGB)
```

```python
# displaying Isolated image

plt.imshow(img_rgb), plt.axis("off")
plt.show()
```

## 8.9 Binarizing the image

Simplified output of an image, Thresholding is setting pixel with intensity greater than some value to be white and less than some value to be black.

In adaptive thresholding the pixel values are determined by pixel intensities of its neighbors.

A major use of thresholding the image is denoising the image.

```python
# Reading the image in gray

image = cv2.imread("birdy.jpg")

image_gray = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
```

```python
# applying adaptive thresholding

max_output_value = 255
neighborhood_size = 11
subtract_from_mean = 2

image_binarized = cv2.adaptiveThreshold(image_gray,
                                       max_output_value,
                                       cv2.ADAPTIVE_THRESH_GAUSSIAN_C,
                                       cv2.THRESH_BINARY,
                                       neighborhood_size,
                                       subtract_from_mean)
```

```python
# displaying the image

plt.imshow(image_binarized, cmap="gray"), plt.axis("off")
plt.show()
```

```python
image_binarized_m = cv2.adaptiveThreshold(image_gray,
                                       max_output_value,
                                       cv2.ADAPTIVE_THRESH_MEAN_C,
                                       cv2.THRESH_BINARY,
                                       neighborhood_size,
                                       subtract_from_mean)
```

```python
# displaying the image

plt.imshow(image_binarized_m, cmap="gray"), plt.axis("off")
plt.show()
```

## 8.10 Removing the Backgrounds

```python
# loading image

image_bgr = cv2.imread("birdy.jpg")

# converting BGR to RGB

image_rgb = cv2.cvtColor(image_bgr, cv2.COLOR_BGR2RGB)
```

```python
# Rectangle values: start x, start y, width, height

rectangle = (0, 56, 256, 150)
```

```python
# creating initial mask

mask  = np.zeros(image_rgb.shape[:2], np.uint8)
```

```python
# creating temporary arrays used by grabCut

bgdModel = np.zeros((1,65), np.float64)
fgdModel = np.zeros((1,65), np.float64)
```

```python
# Run grabCut

cv2.grabCut(image_rgb,
           mask,
           rectangle,
           bgdModel, # temporary array for background
           fgdModel, # temporary array for background
           5, # Number of iterations
           cv2.GC_INIT_WITH_RECT) # Initiative using rectangle
```

```python
# creating masks where sure and likely backgrounds set to 0, otherwise 1

mask2 = np.where((mask == 2) | (mask == 0), 0, 1).astype('uint8')
```

```python
# Multiplying image with new mask to subtract background

image_rgb_nobg = image_rgb * mask2[:, :, np.newaxis]
```

```python
# displaying the image

plt.imshow(image_rgb_nobg), plt.axis("off")
plt.show()
```

GrabCut might not remove few areas of the image background, we can do it manually by marking those areas, but in real world we might be dealing with huge no of images so its not feasible. We accept that part as an some background noise.

```python
# show mask

plt.imshow(mask, cmap='gray'), plt.axis("off")
plt.show()

# GrabCut assumes everything outside the rectangle is the definitely a background 
# and it uses that information and tries to figure out which is actually the background inside the rectangle
```

## 8.11 Detecting Edges

We do this by using edge detection techniques, like Canny edge detector, for detecting intersection of two edges

```python
# Loading grayscale image

image_gray = cv2.imread("birdy.jpg", cv2.IMREAD_GRAYSCALE)
```

```python
# calculating median intensity

median_intensity = np.median(image_gray)
```

```python
# setting thresholds 1 standard deviation above and below median intensity

lower_threshold = int(max(0,(1.0 - 0.33) * median_intensity))
upper_threshold = int(min(255,(1.0 + 0.33)) * median_intensity)
```

```python
# applying canny edge detector

image_canny = cv2.Canny(image_gray, lower_threshold, upper_threshold)
```

```python
# show image

plt.imshow(image_canny, cmap = 'gray'), plt.axis("off")
plt.show()
```

Canny technique uses low and high thresholds, pixels between low and high threshholds are called weak pixels and pixels above high threshold are considered strong pixels.

Sobel filters, Laplacian edge detector are some of the edge detection techniques like Canny edge detector

Edge detection is important to because it has high-information, it allows us to remove low-information areas and focus on high information areas.

## 8.12 Detecting Corners

Detecting corners of an image using Harris Corner detector

```python
# Loading the image and converting it to grayscale

image_bgr = cv2.imread("birdy.jpg")
image_gray = cv2.cvtColor(image_bgr, cv2.COLOR_BGR2GRAY)
image_gray = np.float32(image_gray)
```

```python
# setting corner detector parameters

block_size = 2
aperture = 29
free_parameter = 0.04
```

```python
# Detecting corners

detector_response = cv2.cornerHarris(image_gray,
                                     block_size, # Size of neighbor around each pixel for corner detection
                                     aperture, # Size of sobel kernel
                                     free_parameter) # Larger values corresponds to identifying softer corners
```

```python
# Large corner markers

detector_response = cv2.dilate(detector_response, None)
```

```python
# Detector thresholds greater than threshold, marking them as white.

threshold = 0.02
image_bgr[detector_response > threshold * detector_response.max()] = [255, 255, 255]
```

```python
# converting to grayscale

image_gray = cv2.cvtColor(image_bgr, cv2.COLOR_BGR2GRAY)
```

```python
# display image

plt.imshow(image_gray, cmap = "gray"), plt.axis("off")
plt.show()
```

```python
# potential corners

plt.imshow(detector_response, cmap = "gray"), plt.axis("off")
plt.show()
```

```python
image_bgr = cv2.imread("birdy.jpg")
image_gray = cv2.cvtColor(image_bgr, cv2.COLOR_BGR2GRAY)
```

```python
# no of corners to detect

corners_to_detect = 10
minimum_quality_score = 0.05
minimum_distance = 25
```

```python
# detect corners

corners = cv2.goodFeaturesToTrack(image_gray,
                                 corners_to_detect,
                                 minimum_quality_score,
                                 minimum_distance)

corners = np.float32(corners)
```

```python
# to draw white circle at each corner

for corner in corners:
    x, y = corner[0]
    cv2.circle(image_bgr, (x,y), 10, (255, 255, 255), -1)
```

```python
# Convert to grayscale
image_rgb = cv2.cvtColor(image_bgr, cv2.COLOR_BGR2GRAY)
# Show image
plt.imshow(image_rgb, cmap='gray'), plt.axis("off")
plt.show()
```

## 8.14 Encoding Mean Color as a Feature

## 8.15 Encoding Color Histograms as Features

creating a feature depending the colors in the image.

```python
# loading the image

image_bgr = cv2.imread('birdy.jpg', cv2.IMREAD_COLOR)
```

```python
# converting it to RGB

image_rgb = cv2.cvtColor(image_bgr, cv2.COLOR_BGR2RGB)
```

```python
# creating a list for feature values

features = []
```

```python
# calculate the histogram for each color channel 

colors = ("r", "g", "b")
```

```python
# calculating histogram for each channel and add to feature value list

for  i, channel in enumerate(colors):
    histogram = cv2.calcHist([image_rgb], # image
                            [i], # Index of channel
                            None, # No mask
                            [256], # Histogram size
                            [0, 256]) # range
    features.extend(histogram)
```

```python
# creating a vector for observation feature values

observation = np.array(features).flatten()
```

```python
# first five observation values

observation[0:5]
```

```python
# each color in the image is a combination of red, green and blue colors

image_rgb[0,0]
```
