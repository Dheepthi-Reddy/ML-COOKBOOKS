## 8.1 Loading Images

```python
import numpy as np
from matplotlib import pyplot as plt
```

```python
# loading image as grayscale

image = cv2.imread("birdy.jpg", cv2.IMREAD_GRAYSCALE)
```

```python
# displaying the image with matplotlib

plt.imshow(image, cmap = "gray"), plt.axis("off")
plt.show()
```

Image is a data by using imread we are converting image data into a numpy array(Image into a matrix). Each element represnts a pixel.

```python
# each element contains 3 values corresponding to Blue, Green, Red(BGR):

# loading image in color

img = cv2.imread("birdy.jpg", cv2.IMREAD_COLOR)

# display image

plt.imshow(img), plt.axis("off")
plt.show()
```

```python
# show pixel

img[0,0]
```

OpenCV uses BGR format whereas Matplotlib uses RGB, Blue and Red clors are swapped, inorder to display the colors properly we need to convert BGR to RGB

```python
# Converting BGR to RGB

img_rgb = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)

# display image

plt.imshow(img_rgb), plt.axis("off")
plt.show()
```

## 8.2 Saving images

Saving an image for preprocessing

```python
# loading image

image_save = cv2.imread("birdy.jpg", cv2.IMREAD_GRAYSCALE)

# saving the image

cv2.imwrite("birdy.jpg", image_save)
```

"imwrite" saves the image to specified filepath, "imwrite" overwrite the existing image without outputting any error

## 8.3 Resizing the Imgaes

Resizing the image for preprocessing

```python
# laoding the image by grayscale

image_resize = cv2.imread("birdy.jpg", cv2.IMREAD_GRAYSCALE)

# resizing the image to 50x50 pixels

resized_img = cv2.resize(image_resize, (50,50))
```

```python
# display the image

plt.imshow(resized_img, cmap="gray"), plt.axis("off")

plt.show()
```

1) Images come in all sizes to use images as features they all need to be in same size.
2) ML requires thousands of images, some images can be of large size taking up too much memory, by resizing them we can save memory.

some common sizes used in ML are 32x32, 64x64, 96x96, 256x256

## 8.4 Cropping Images

images are read as numpy arrays and matrices, we can do array slicing.

```python
# loading image in grayscale

img = cv2.imread("birdy.jpg", cv2.IMREAD_GRAYSCALE)
```

```python
# selecting few columns and all the rows

img_crop = img[:,90:200]
```

```python
# displaying the image

plt.imshow(img_crop, cmap = "gray"), plt.axis("off")
plt.show()
```
